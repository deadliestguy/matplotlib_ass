```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
plt.scatter( x , y ,marker='v',c = colours, alpha=0.5 )
plt.xlabel("x-->")
plt.ylabel("y-->")
plt.title('scatter plot')
```
