```python
import plotly.graph_objects as go
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 1. x axis = age and y axis = fare scatter plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic['age'] , y=titanic['fare'] , mode='markers' ,marker_color='green',marker_size = 10,opacity=0.5 ))
```

```python
# q.02-> tips dataset er box plot
tips = sns.load_dataset('tips')
tips.head()
fig = go.Figure()
fig.add_trace(go.Box(tips))
```

```python
# q.03 - > histogram
# of sex colm
go.Figure(data=[go.Histogram(x=tips['sex'],color = tips['day'])])
```

```python
# histogram of total bill
go.Figure(data=[go.Histogram(y=tips['total_bill'])])
```
